```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look better
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("students_data.csv")
print(df.head())
```

```python
# Plot student marks with a line plot
plt.figure(figsize=(8, 4))
plt.plot(df["Name"], df["Marks"], marker='o', linestyle='-', color='blue')
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df["Name"], df["Marks"], color='skyblue')
plt.title("Marks Bar Chart")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Subject", data=df, palette="Set2")
plt.title("Subject Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["Marks"], bins=5, kde=True, color='orange')
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x="Subject", y="Marks", data=df, palette="coolwarm")
plt.title("Marks Distribution by Subject")
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```
